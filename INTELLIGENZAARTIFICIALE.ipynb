```python
#BLOCCO 9

from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn import metrics
import time
from datetime import timedelta
start_time = time.monotonic()

x_graph=np.zeros(9)
y_train_graph=np.zeros(9)
y_test_graph=np.zeros(9)


for k in range(10,19):
    if(k!=18):
        dim=2**k
        
    else:
        dim=138610
    
    print("DIMENSIONE TRAINING SET= ",dim)
    Xtrain_considered = X_train[:dim,:]
    ytrain_considered = y_train[:dim]
     
    print("X_train.shape= ",Xtrain_considered.shape)
    print("y_train.shape= ",ytrain_considered.shape)
    print("X_test.shape= ",X_test.shape)
    print("y_test.shape= ",y_test.shape)
    reg=RandomForestClassifier(n_estimators=10, max_depth=None,min_samples_split=2)
    
    reg.fit(Xtrain_considered,ytrain_considered)
    y_test_pred=reg.predict(X_test)
    y_train_pred=reg.predict(Xtrain_considered)
    print(f'Accuracy_test:  {metrics.accuracy_score(y_test, y_test_pred)*100:.4}%')
    print(f'Accuracy_train:  {metrics.accuracy_score(ytrain_considered, y_train_pred)*100:.4}%')
    accuracy_test= float(metrics.accuracy_score(y_test, y_test_pred)*100)
    accuracy_train=float(metrics.accuracy_score(ytrain_considered, y_train_pred)*100)
    error_train=float((1-reg.score(Xtrain_considered,ytrain_considered))*100)
    error_test=float((1-reg.score(X_test,y_test))*100)
    

    print(f'Errore di predizione sul test: {(1-reg.score(X_test,y_test))*100:.4}%')
    print(f'Errore di predizione sul train: {(1-reg.score(Xtrain_considered,ytrain_considered))*100:.4}%')
        
    x_graph[k-10]=int(dim)
    y_test_graph[k-10]=error_test
    y_train_graph[k-10]=error_train
    print()
    
    plt.plot(x_graph[:k-9], y_test_graph[:k-9], 'ro')
    plt.plot(x_graph[:k-9], y_test_graph[:k-9], "r-")
    plt.grid()
    plt.title("Andamento dell'errore di predizione sul test set")
    plt.xlabel('Dimensione training set')
    plt.ylabel('Errore in percentuale')
    plt.show()
    
    plt.plot(x_graph[:k-9], y_train_graph[:k-9], 'bo')
    plt.plot(x_graph[:k-9], y_train_graph[:k-9], "b-")
    plt.grid()
    plt.title("Andamento dell'errore di predizione sul training set")
    plt.xlabel('Dimensione training set')
    plt.ylabel('Errore in percentuale')
    plt.show()
    print("--------------------------------------------------------------------------------------------------------")
    print()

plt.title("Andamento dell'errore di predizione")
plt.xlabel('Dimensione training set')
plt.ylabel('Errore in percentuale')
plt.plot(x_graph, y_test_graph, 'r')
plt.plot(x_graph, y_train_graph, 'b')
plt.grid()
plt.show()

end_time = time.monotonic()
print(timedelta(seconds=end_time - start_time))
```
